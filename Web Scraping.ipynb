Steps involved in web scraping:

-- Send an HTTP request to the url of the webpage you want to access.
    The server responds to the request by returning the HTML content of the webpage.
    Library used--requests.
--After accessing the HTML content, parsing the data.
   Since most of the HTML data is nested, we cannot extract data simply through
   string processing.
   One needs a parser which can create a nested/tree structure of the HTML data.
   library used--html5lib.
--Fianlly, we need to search  the parse tree that we created, i.e. tree traversal.
   Library used--  Beautiful Soup.



```python
#Accessing the HTML content from webpage using request module
import requests
from bs4 import BeautifulSoup#very famous scrapping
import html5lib#helping bs understand the html content by creating a tree structure
url= "https://shirish137.github.io/shirishonline/" #specify the URL of the webpage
res= requests.get(url)#this is a fn from requests library
print(res)
print(res.content)#content can be used to get the content from our res object
```

```python
soup= BeautifulSoup(res.content, 'html5lib') 
print(soup)
```

```python
print(soup.prettify())#optional to make the content good looking by ordering removing spaces
```

#Searching and navigating through the parse tree
'''
--To extract some useful data from the HTML content using using find() or  findAll() method
--We can also save all relevant information.

--find(): returns the first matching element
--find_all(): returns a list of all matching elements [by id, class, or tag name]
'''

```python

data=soup.find_all('p')  # filters only paragraphs(tagname)
#find all return all matches as a list
print(data[0])
print(data[1])
```

```python
fdata=soup.find('p')#first match
print(fdata)
```

```python
data=soup.find_all(id='name')  #using id
#id in html is used to differentiate one element from other in html
#using we apply style,js and all in web designing
print(data)
```

```python
data=soup.find_all(class_='card-text')  # using class name
#class is used in html to group elements and identify them together
print(data[0])
```

```python
print(data[0].get_text())
```

```python
data=soup.find_all('p')  # filters only paragraphs(tagname)
print(data[0])
print(data[0].get_text())
```

```python
print(data)
for i in data:
    print(i.get_text())
```

```python
#Removing Rs to get only numeric values so that we can perform ops
import re
data=soup.find_all(class_='card-text')
l=[]
for i in data:
    t=i.get_text()
    #print(t)
    li=re.sub('Rs.',' ',t)
    #print(li)
    l.append(int(li))
print(l)
```

```python
#I can make assumptions from the data
print("Total of all items on the page: ",sum(l))
print("Avg price of items: ",sum(l)/len(l))
```
